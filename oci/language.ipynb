```python
import os
from dotenv import load_dotenv
from oci.config import from_file
from oci.ai_language.ai_service_language_client import AIServiceLanguageClient
```

```python
config = from_file()
client = AIServiceLanguageClient(config=config)
```

```python
from oci.ai_language.models import BatchDetectLanguageSentimentsDetails, TextDocument
```

```python
_ = load_dotenv()
compartment_id = os.getenv("COMPARTMENT_ID")
endpoint = os.getenv("LANGUAGE_ENDPOINT")
```

```python
import pandas as pd
import uuid
```

```python
documents = []
```

```python

df = pd.read_csv("./JapaneseNLP.csv", encoding="utf-8")
df.head()
```

```python
for idx, row in df.iterrows():
    documents.append(TextDocument(
        key=str(uuid.uuid4()),
        text=row["Japan"],
        language_code="ja"
    ))
    if len(documents) == 100:
        response = client.batch_detect_language_sentiments(
            batch_detect_language_sentiments_details=BatchDetectLanguageSentimentsDetails(
                compartment_id=compartment_id,
                documents=documents,
            )
        )
        print(response.data)
        documents = []
```
